```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
#Preprocessing
df = df[df['21'] != -99.999]
df = df[df['21'] < 22]
df = df[df['33'] > 0.9]

#Calculating Luminosities
L = np.power(10, -0.4*df[df['21'] != -99.999]['21'])
```

```python
#converting into cartesian coordinates

def cartesian(dec,RA,r):
    x=r*np.sin(dec)*np.cos(RA)
    y=r*np.sin(dec)*np.sin(RA)
    z=r*np.cos(dec)
    return (x,y,z)
dec=df['35']
RA=df['34']
r=10.4
x, y, z = cartesian(dec,RA,r)


#assuming spherical system, calculating the origin
x_center = (x.min()+x.max())/2
y_center = (y.min()+y.max())/2
z_center = (z.min()+z.max())/2


#shifting the origin
x1 = x - x_center
y1 = y - y_center
z1 = z - z_center
```

```python
#assiging the cartesian cooridinates to the dataframe

df['x1'] = x1
df['y1'] = y1
df['z1'] = z1
```

```python
#Scaling up the coordinates for better accuracy in result

df['x1'] = df['x1']*(10**3)
df['y1'] = df['y1']*(10**3)
df['z1'] = df['z1']*(10**3)
df['L'] = L
```

```python
#getting the limits of the for iteration
ma = np.sum(df['L'])
x_max = np.abs(df['x1']).max()
y_max = np.abs(df['y1']).max()
z_max = np.abs(df['z1']).max()

x_min = np.abs(df['x1']).min()
y_min = np.abs(df['y1']).min()
z_min = np.abs(df['z1']).min()

hi = np.max([x_max, y_max, z_max])
lo = np.min([x_min, y_min, z_min])
hi = np.round(hi)
lo = np.round(lo)
hi = int(hi)
lo = int(lo)
print(hi)
print(lo)
```

```python
#Calculating the the radius

cum = 0
for i in range(lo, hi):
    cum = np.sum(df[(df['x1']**2 + df['y1']**2 + df['z1']**2) < i**2]['L'])
    if(cum > ma/2):
        print(i)
        break;
print("Half-Light Radius is:")
print(i*3.262)
```
